```python
# ============================================================
# PMI Components -> GDP Forecasting (Full Pipeline, FIXED)
# ============================================================
# Key Fix:
# - Never plot pd.Period directly (Matplotlib cannot convert Period -> float).
# - All time series plots use x = 0..n-1 and label ticks with "YYYYQn" strings.
#
# Implements:
# (1) Predictive regressions with HAC (Newey–West) SEs
# (2) Expanding-window pseudo out-of-sample forecasting
# (3) Forecast comparison (RMSFE/MAE/ME) + DM-style HAC test on loss differences
# (4) Shapley-value decomposition of in-sample R^2 for correlated predictors
# (5) VIF multicollinearity diagnostics
#
# Outputs:
# - MANY plots shown with plt.show()
# - MANY tables displayed with display()
# - Everything auto-saved + ZIP exported
# ============================================================

# ---------------------------
# 0) Install (Colab-friendly)
# ---------------------------
import sys, subprocess
def _pip_install(pkg):
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", pkg])

for pkg in ["openpyxl", "statsmodels"]:
    try:
        __import__(pkg if pkg != "openpyxl" else "openpyxl")
    except Exception:
        _pip_install(pkg)

# ---------------------------
# 1) Imports
# ---------------------------
import os, re, math, itertools, shutil, zipfile, datetime
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from IPython.display import display, Markdown

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 160)

# ---------------------------
# 2) File paths
# ---------------------------
FILE_PMI_FINAL = "PMI-Final data.xlsx"
FILE_RBI       = "PMI-SS(RBI).xlsx"

# ---------------------------
# 3) Output directory
# ---------------------------
RUNSTAMP = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
OUTROOT  = Path(f"PMI_GDP_Forecast_Run_{RUNSTAMP}")
FIGDIR   = OUTROOT / "figures"
TABDIR   = OUTROOT / "tables"
REPDIR   = OUTROOT / "reports"
DATADIR  = OUTROOT / "data_clean"

for d in [FIGDIR, TABDIR, REPDIR, DATADIR]:
    d.mkdir(parents=True, exist_ok=True)

print("Outputs will be saved to:", OUTROOT.resolve())

# ---------------------------
# 4) Helper functions
# ---------------------------
def save_df(df: pd.DataFrame, name: str, index=False):
    """Save table as CSV + LaTeX and display it."""
    csv_path = TABDIR / f"{name}.csv"
    tex_path = TABDIR / f"{name}.tex"
    df.to_csv(csv_path, index=index)
    try:
        df.to_latex(
            tex_path,
            index=index,
            float_format=lambda x: f"{x:.4f}" if isinstance(x, (int,float,np.floating)) else str(x)
        )
    except Exception:
        pass
    display(Markdown(f"### Table: {name}"))
    display(df)

def save_fig(fig, name: str, dpi=180):
    """Save figure as PNG + PDF, then show it."""
    png_path = FIGDIR / f"{name}.png"
    pdf_path = FIGDIR / f"{name}.pdf"
    fig.savefig(png_path, dpi=dpi, bbox_inches="tight")
    fig.savefig(pdf_path, bbox_inches="tight")
    plt.show()
    plt.close(fig)

def period_from_quarter_string(q: str):
    # expects 'YYYY-Qn' -> Period('YYYYQn', 'Q')
    return pd.Period(q.replace("-", ""), freq="Q")

def clean_quarter_df(df: pd.DataFrame, date_col="Date"):
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]

    df = df[df[date_col].astype(str).str.match(r"^\d{4}-Q[1-4]$")]
    df["period"] = df[date_col].astype(str).apply(period_from_quarter_string)
    df = df.sort_values("period").reset_index(drop=True)

    for c in df.columns:
        if c in [date_col, "period"]:
            continue
        df[c] = pd.to_numeric(df[c], errors="coerce")
    return df

def nw_lags(T):
    L = int(np.floor(4 * (T / 100) ** (2 / 9)))
    return max(1, min(8, L))

def fit_hac_ols(y: pd.Series, X: pd.DataFrame, lags=None):
    Xc = sm.add_constant(X, has_constant="add")
    model = sm.OLS(y, Xc, missing="drop")
    if lags is None:
        lags = nw_lags(len(y))
    res = model.fit(cov_type="HAC", cov_kwds={"maxlags": lags})
    return res

def dm_hac_test(loss_model, loss_base, hac_lags=4):
    """
    DM-style test: d_t = loss_model - loss_base
    Test E[d_t]=0 using OLS(d ~ const) with HAC SE.
    """
    d = pd.Series(loss_model).reset_index(drop=True) - pd.Series(loss_base).reset_index(drop=True)
    d = d.dropna().reset_index(drop=True)
    X = np.ones((len(d), 1))
    res = sm.OLS(d.values, X).fit(cov_type="HAC", cov_kwds={"maxlags": hac_lags})
    return {
        "mean_diff": float(d.mean()),
        "t_stat": float(res.tvalues[0]),
        "p_value": float(res.pvalues[0]),
        "n": int(len(d))
    }

def descriptive_stats(df, cols):
    out = df[cols].describe().T
    out["skew"] = df[cols].skew(numeric_only=True)
    out["kurt"] = df[cols].kurt(numeric_only=True)
    return out

def time_axis(df_period_series):
    """
    Returns:
      x: numeric axis 0..n-1
      labels: 'YYYYQn' strings
      tick_idx: chosen tick positions
      tick_labels: corresponding labels
    """
    labels = df_period_series.astype(str).tolist()
    n = len(labels)
    x = np.arange(n)

    # show about 10 ticks max
    step = max(1, n // 10)
    tick_idx = np.arange(0, n, step)
    tick_labels = [labels[i] for i in tick_idx]
    return x, labels, tick_idx, tick_labels

def corr_heatmap(df, cols, title, fig_name):
    C = df[cols].corr()
    fig, ax = plt.subplots(figsize=(10, 8))
    im = ax.imshow(C.values, aspect="auto")
    ax.set_xticks(range(len(cols)))
    ax.set_yticks(range(len(cols)))
    ax.set_xticklabels(cols, rotation=45, ha="right")
    ax.set_yticklabels(cols)
    ax.set_title(title)
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    save_fig(fig, fig_name)
    return C

def plot_series(df, ycols, title, fig_name):
    x, labels, tick_idx, tick_labels = time_axis(df["period"])
    fig, ax = plt.subplots(figsize=(12, 5))
    for c in ycols:
        ax.plot(x, df[c].values, linewidth=2, label=c)
    ax.set_title(title)
    ax.set_xlabel("Quarter")
    ax.set_xticks(tick_idx)
    ax.set_xticklabels(tick_labels, rotation=45, ha="right")
    ax.grid(True, alpha=0.3)
    ax.legend(ncol=2, fontsize=9)
    save_fig(fig, fig_name)

def plot_single(df, ycol, title, fig_name):
    x, labels, tick_idx, tick_labels = time_axis(df["period"])
    fig, ax = plt.subplots(figsize=(12, 4))
    ax.plot(x, df[ycol].values, linewidth=2)
    ax.set_title(title)
    ax.set_xlabel("Quarter")
    ax.set_ylabel(ycol)
    ax.set_xticks(tick_idx)
    ax.set_xticklabels(tick_labels, rotation=45, ha="right")
    ax.grid(True, alpha=0.3)
    save_fig(fig, fig_name)

def scatter_with_fit(df, xcol, ycol, title, fig_name):
    xdat = df[xcol].values
    ydat = df[ycol].values
    ok = np.isfinite(xdat) & np.isfinite(ydat)
    xdat = xdat[ok]; ydat = ydat[ok]

    X = sm.add_constant(xdat)
    res = sm.OLS(ydat, X).fit()
    xs = np.linspace(xdat.min(), xdat.max(), 200)
    ys = res.predict(sm.add_constant(xs))

    fig, ax = plt.subplots(figsize=(6.8, 5))
    ax.scatter(xdat, ydat, alpha=0.75)
    ax.plot(xs, ys, linewidth=2)
    ax.set_title(title)
    ax.set_xlabel(xcol)
    ax.set_ylabel(ycol)
    ax.grid(True, alpha=0.3)
    save_fig(fig, fig_name)
    return res

def compute_vif(dfX):
    """VIF table for multicollinearity diagnostics."""
    X = dfX.copy()
    X = X.replace([np.inf, -np.inf], np.nan).dropna()
    X = sm.add_constant(X, has_constant="add")
    vifs = []
    for i, col in enumerate(X.columns):
        if col == "const":
            continue
        vifs.append({
            "variable": col,
            "VIF": float(variance_inflation_factor(X.values, i))
        })
    return pd.DataFrame(vifs).sort_values("VIF", ascending=False).reset_index(drop=True)

# ---------------------------
# 5) Core methodology
# ---------------------------
def component_regressions(dfm, y_col, comp_cols, ar_col="y_lag", covid_col="covid", hac_lags=None):
    rows = []
    for comp in comp_cols:
        X = pd.DataFrame({
            "y_lag": dfm[ar_col],
            "covid": dfm[covid_col],
            "x": dfm[comp]
        })
        res = fit_hac_ols(dfm[y_col], X, lags=hac_lags)
        rows.append({
            "predictor": comp,
            "beta_x": res.params.get("x", np.nan),
            "se_x": res.bse.get("x", np.nan),
            "t_x": res.tvalues.get("x", np.nan),
            "p_x": res.pvalues.get("x", np.nan),
            "R2": res.rsquared,
            "Adj_R2": res.rsquared_adj,
            "N": int(res.nobs)
        })
    out = pd.DataFrame(rows).sort_values("p_x").reset_index(drop=True)
    out["rank_p"] = np.arange(1, len(out) + 1)
    return out

def joint_regression(dfm, y_col, comp_cols, ar_col="y_lag", covid_col="covid", hac_lags=None):
    X = pd.DataFrame({"y_lag": dfm[ar_col], "covid": dfm[covid_col]})
    for c in comp_cols:
        X[c] = dfm[c]
    res = fit_hac_ols(dfm[y_col], X, lags=hac_lags)
    return res

def expanding_oos_forecast(dfm, y_col, predictors, start=20):
    """
    Expanding-window:
      train [0..t-1], predict t
    """
    n = len(dfm)
    yhat = np.full(n, np.nan)

    for t in range(start, n):
        train = dfm.iloc[:t]
        test  = dfm.iloc[[t]]

        X_train = sm.add_constant(train[predictors], has_constant="add")
        X_test  = sm.add_constant(test[predictors], has_constant="add")

        res = sm.OLS(train[y_col], X_train, missing="drop").fit()
        yhat[t] = float(res.predict(X_test).iloc[0])

    out = dfm[["Date", "period"]].copy()
    out["y"] = dfm[y_col].values
    out["yhat"] = yhat
    out = out.dropna(subset=["yhat"]).reset_index(drop=True)
    out["err"] = out["y"] - out["yhat"]
    out["sqerr"] = out["err"]**2
    out["abserr"] = out["err"].abs()
    return out

def forecast_metrics(fc):
    e = fc["err"].values
    return {
        "N_oos": int(len(e)),
        "RMSFE": float(np.sqrt(np.mean(e**2))),
        "MAE": float(np.mean(np.abs(e))),
        "ME": float(np.mean(e))
    }

def shapley_r2(dfm, y_col, base_cols, comp_cols):
    """
    Shapley decomposition of in-sample R^2 over comp_cols,
    keeping base_cols always included.
    """
    y = dfm[y_col]
    K = len(comp_cols)
    r2_map = {}

    def r2_for_subset(S):
        cols = base_cols + list(S)
        X = sm.add_constant(dfm[cols], has_constant="add")
        res = sm.OLS(y, X, missing="drop").fit()
        return float(res.rsquared)

    for r in range(K + 1):
        for subset in itertools.combinations(comp_cols, r):
            r2_map[frozenset(subset)] = r2_for_subset(subset)

    phi = {c: 0.0 for c in comp_cols}
    for c in comp_cols:
        others = [o for o in comp_cols if o != c]
        for r in range(K):
            for subset in itertools.combinations(others, r):
                S = frozenset(subset)
                S_with = S | {c}
                w = math.factorial(r) * math.factorial(K - r - 1) / math.factorial(K)
                phi[c] += w * (r2_map[S_with] - r2_map[S])

    out = pd.DataFrame({"predictor": list(phi.keys()), "shapley_R2": list(phi.values())})
    out = out.sort_values("shapley_R2", ascending=False).reset_index(drop=True)
    return out, r2_map[frozenset()], r2_map[frozenset(comp_cols)]

# ---------------------------
# 6) Load datasets
# ---------------------------
def load_pmi_final(path):
    df = pd.read_excel(path, sheet_name="Data Sheet")
    df = clean_quarter_df(df, date_col="Date")

    # create lead GDP and "real_proxy" as in your sheet
    df["gdp_lead"] = df["GDP Growth"].shift(-1)
    df["infl"] = df["CPI (Q-O-Q)"] if "CPI (Q-O-Q)" in df.columns else np.nan
    df["real_proxy"] = df["gdp_lead"] - df["infl"]

    # lags
    df["y_lag"] = df["GDP Growth"]
    df["covid"] = df["Covid dummy"] if "Covid dummy" in df.columns else 0.0

    df.to_csv(DATADIR / "PMI_Final_clean.csv", index=False)
    return df

def load_rbi(path):
    df = pd.read_excel(path, sheet_name="Data2")
    df = clean_quarter_df(df, date_col="Date")
    df["covid"] = df["Covid dummy"] if "Covid dummy" in df.columns else 0.0
    if "GDP Growth" in df.columns:
        df["y_lag"] = df["GDP Growth"].shift(1)
    df.to_csv(DATADIR / "RBI_Data2_clean.csv", index=False)
    return df

# ---------------------------
# 7) Full analysis driver
# ---------------------------
def analyze_dataset(df, dataset_name, target_col, comp_cols, start_oos=20):
    print("\n" + "="*95)
    print(f"DATASET: {dataset_name} | TARGET: {target_col}")
    print("="*95)

    # build modeling frame
    keep = ["Date", "period", target_col, "y_lag", "covid"] + comp_cols
    dfm = df[keep].copy()
    dfm = dfm.dropna(subset=[target_col, "y_lag"] + comp_cols).reset_index(drop=True)

    dfm.to_csv(DATADIR / f"{dataset_name}_{target_col}_model_frame.csv", index=False)

    display(Markdown(f"## {dataset_name}: modeling frame preview ({target_col})"))
    display(dfm.head(8))

    # descriptive stats
    desc = descriptive_stats(dfm, [target_col, "y_lag", "covid"] + comp_cols)
    save_df(desc.round(4), f"{dataset_name}_{target_col}_descriptive_stats", index=True)

    # correlations
    C = corr_heatmap(dfm, [target_col] + comp_cols,
                     title=f"{dataset_name}: Correlations (target + PMI components)",
                     fig_name=f"{dataset_name}_{target_col}_corr_heatmap")
    save_df(C.round(4), f"{dataset_name}_{target_col}_correlation_matrix", index=True)

    # time plots (fixed)
    plot_single(dfm, target_col, f"{dataset_name}: {target_col} over time", f"{dataset_name}_{target_col}_ts_target")
    plot_series(dfm, comp_cols, f"{dataset_name}: PMI components over time", f"{dataset_name}_{target_col}_ts_components")

    # scatter plots each component vs target
    for c in comp_cols:
        safe = re.sub(r"[^A-Za-z0-9]+", "_", c)
        scatter_with_fit(dfm, c, target_col,
                         title=f"{dataset_name}: {target_col} vs {c} (scatter + OLS line)",
                         fig_name=f"{dataset_name}_{target_col}_scatter_{safe}")

    # HAC regressions
    hacL = nw_lags(len(dfm))
    comp_reg = component_regressions(dfm, target_col, comp_cols, hac_lags=hacL)
    save_df(comp_reg.round(4), f"{dataset_name}_{target_col}_component_regressions_HAC", index=False)

    # VIF
    vif = compute_vif(dfm[comp_cols])
    save_df(vif.round(4), f"{dataset_name}_{target_col}_VIF_components", index=False)

    # joint regression HAC
    joint = joint_regression(dfm, target_col, comp_cols, hac_lags=hacL)
    summ_path = REPDIR / f"{dataset_name}_{target_col}_joint_regression_summary.txt"
    summ_path.write_text(joint.summary().as_text(), encoding="utf-8")

    joint_tab = pd.DataFrame({
        "coef": joint.params,
        "se(HAC)": joint.bse,
        "t(HAC)": joint.tvalues,
        "p(HAC)": joint.pvalues
    }).round(4)
    save_df(joint_tab, f"{dataset_name}_{target_col}_joint_regression_coefs_HAC", index=True)

    # ---------- OOS Forecasts ----------
    models = {}
    models["Baseline_AR+Covid"] = ["y_lag", "covid"]

    for c in comp_cols:
        models[f"AR+Covid+{c}"] = ["y_lag", "covid", c]

    models["Joint_AllComponents"] = ["y_lag", "covid"] + comp_cols

    forecasts = {}
    met_rows = []
    for name, preds in models.items():
        fc = expanding_oos_forecast(dfm, target_col, preds, start=start_oos)
        forecasts[name] = fc
        m = forecast_metrics(fc)
        m["model"] = name
        met_rows.append(m)

    oos_tab = pd.DataFrame(met_rows).sort_values("RMSFE").reset_index(drop=True)
    save_df(oos_tab.round(4), f"{dataset_name}_{target_col}_OOS_metrics", index=False)

    # RMSFE bar plot (fixed)
    fig, ax = plt.subplots(figsize=(12, 5))
    ax.bar(oos_tab["model"], oos_tab["RMSFE"])
    ax.set_title(f"{dataset_name}: OOS RMSFE by model (target={target_col})")
    ax.set_ylabel("RMSFE")
    ax.tick_params(axis="x", rotation=75)
    ax.grid(True, axis="y", alpha=0.3)
    save_fig(fig, f"{dataset_name}_{target_col}_bar_RMSFE")

    # Actual vs predicted for top 2 models (fixed)
    best2 = list(oos_tab["model"].iloc[:2])
    base_fc = forecasts[best2[0]].copy()
    x, labels, tick_idx, tick_labels = time_axis(base_fc["period"])

    fig, ax = plt.subplots(figsize=(12, 5))
    ax.plot(x, base_fc["y"].values, linewidth=2, label="Actual")
    for bm in best2:
        fc = forecasts[bm]
        ax.plot(np.arange(len(fc)), fc["yhat"].values, linewidth=2, label=f"Pred: {bm}")
    ax.set_title(f"{dataset_name}: Actual vs Predicted ({target_col}) — top 2 OOS models")
    ax.set_xticks(tick_idx)
    ax.set_xticklabels(tick_labels, rotation=45, ha="right")
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=9)
    save_fig(fig, f"{dataset_name}_{target_col}_pred_vs_actual_top2")

    # Residual diagnostics for best model
    best_fc = forecasts[oos_tab.loc[0, "model"]]
    fig, ax = plt.subplots(figsize=(7, 4.5))
    ax.hist(best_fc["err"].values, bins=12)
    ax.set_title(f"{dataset_name}: OOS forecast errors histogram (best model)")
    ax.set_xlabel("Error")
    ax.set_ylabel("Count")
    ax.grid(True, alpha=0.3)
    save_fig(fig, f"{dataset_name}_{target_col}_best_errors_hist")

    # Cumulative SSE plot across models (nice comparison)
    fig, ax = plt.subplots(figsize=(12, 5))
    for name, fc in forecasts.items():
        ax.plot(np.arange(len(fc)), np.cumsum(fc["sqerr"].values), label=name, linewidth=2)
    ax.set_title(f"{dataset_name}: Cumulative Squared Forecast Error (lower is better)")
    ax.set_xlabel("OOS time index")
    ax.set_ylabel("Cumulative SSE")
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=7, ncol=2)
    save_fig(fig, f"{dataset_name}_{target_col}_cumSSE_all_models")

    # ---------- DM-style HAC tests vs baseline ----------
    baseline_loss = forecasts["Baseline_AR+Covid"]["sqerr"].values
    dm_rows = []
    for name, fc in forecasts.items():
        if name == "Baseline_AR+Covid":
            continue
        test = dm_hac_test(fc["sqerr"].values, baseline_loss, hac_lags=4)
        dm_rows.append({
            "model": name,
            "mean_sqerr_diff(model - baseline)": test["mean_diff"],
            "DM_HAC_t": test["t_stat"],
            "p_value": test["p_value"],
            "N": test["n"]
        })
    dm_tab = pd.DataFrame(dm_rows).sort_values("p_value").reset_index(drop=True)
    save_df(dm_tab.round(4), f"{dataset_name}_{target_col}_DM_tests_vs_baseline", index=False)

    # ---------- Shapley decomposition ----------
    base_cols = ["y_lag", "covid"]
    shap, r2_base, r2_full = shapley_r2(dfm, target_col, base_cols, comp_cols)
    save_df(shap.round(6), f"{dataset_name}_{target_col}_Shapley_R2", index=False)

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.bar(shap["predictor"], shap["shapley_R2"])
    ax.set_title(f"{dataset_name}: Shapley contributions to in-sample R^2 (base={r2_base:.3f}, full={r2_full:.3f})")
    ax.set_ylabel("Shapley R^2 contribution")
    ax.tick_params(axis="x", rotation=45)
    ax.grid(True, axis="y", alpha=0.3)
    save_fig(fig, f"{dataset_name}_{target_col}_bar_ShapleyR2")

    # ---------- compact markdown report ----------
    report = []
    report.append(f"# Report: {dataset_name} | target={target_col}")
    report.append(f"- Rows used: {len(dfm)}")
    report.append(f"- HAC maxlags used: {hacL}")
    report.append(f"- OOS training length: {start_oos}")
    report.append("")
    report.append("## Best OOS models by RMSFE")
    report.append(oos_tab.head(10).to_markdown(index=False))
    report.append("")
    report.append("## Component regressions (ranked by p-value)")
    report.append(comp_reg.head(10).to_markdown(index=False))
    report.append("")
    report.append("## VIF (multicollinearity)")
    report.append(vif.to_markdown(index=False))
    report.append("")
    report.append("## Shapley R^2 attribution")
    report.append(shap.to_markdown(index=False))

    rep_path = REPDIR / f"{dataset_name}_{target_col}_report.md"
    rep_path.write_text("\n".join(report), encoding="utf-8")

    print(f"\nSaved joint regression summary to: {summ_path}")
    print(f"Saved report to: {rep_path}")

# ---------------------------
# 8) RUN: PMI-Final dataset
# ---------------------------
if not os.path.exists(FILE_PMI_FINAL):
    raise FileNotFoundError(f"Cannot find '{FILE_PMI_FINAL}'. Upload it or set the correct path.")

df_main = load_pmi_final(FILE_PMI_FINAL)
display(Markdown("## Loaded: PMI-Final (Data Sheet)"))
display(df_main.head(8))
print("Shape:", df_main.shape)
print("Date range:", df_main["Date"].iloc[0], "to", df_main["Date"].iloc[-1])

# Use deviation-from-50 columns (already present in your sheet)
components = ["Future Output0", "Output0", "New Orders0", "Employment0", "Input Prices0", "Output Prices0"]
missing = [c for c in components if c not in df_main.columns]
if missing:
    raise ValueError(f"Missing expected component columns in PMI-Final: {missing}")

# Run for real_proxy (your paper's main target: gdp_{t+1} - inflation_t)
analyze_dataset(
    df=df_main.copy(),
    dataset_name="PMI_Final",
    target_col="real_proxy",
    comp_cols=components,
    start_oos=20
)

# Run for nominal future GDP growth as robustness (gdp_{t+1})
analyze_dataset(
    df=df_main.copy(),
    dataset_name="PMI_Final",
    target_col="gdp_lead",
    comp_cols=components,
    start_oos=20
)

# ---------------------------
# 9) Optional: RBI dataset robustness
# ---------------------------
if os.path.exists(FILE_RBI):
    df_rbi = load_rbi(FILE_RBI)
    display(Markdown("## Loaded: PMI-SS(RBI) (Data2)"))
    display(df_rbi.head(8))

    # Try likely component columns (keep only those that exist)
    rbi_candidates = [
        "COMPOSITE OUTPUT INDEX( 0)",
        "Manufacturing Index (0)",
        "Employment(0)",
        "Input Prices(0)",
        "Output Prices(0)"
    ]
    rbi_components = [c for c in rbi_candidates if c in df_rbi.columns]

    if len(rbi_components) >= 2 and "GDP Growth" in df_rbi.columns:
        analyze_dataset(
            df=df_rbi.copy(),
            dataset_name="RBI_Data2",
            target_col="GDP Growth",
            comp_cols=rbi_components,
            start_oos=20
        )
    else:
        print("RBI_Data2: Not enough matching columns found for the full pipeline. Skipping.")
else:
    print("RBI file not found; skipping RBI robustness.")

# ---------------------------
# 10) ZIP everything
# ---------------------------
zip_path = shutil.make_archive(str(OUTROOT), "zip", root_dir=str(OUTROOT))
print("\n✅ DONE. Zipped outputs to:")
print(zip_path)
```
